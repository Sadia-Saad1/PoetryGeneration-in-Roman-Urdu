# Extracting First and Last Words of all the Stanzas

The firstword() function extracts the first words of all the verses in poetry.txt.
The lastword() function  extracts the last words of all the verses in poetry.txt.

```python

#Picks first words from the all the sentences in the text.

def firstword(selectsentence):
  add = 0
  firstlist = list()
  for i in range (len(selectsentence)-1):
    words = ""
    if selectsentence[i] != '':
      for j in range(len(selectsentence[i])-1):
        if (selectsentence[i][j] == " "):
          break
        else:
          words += selectsentence[i][j]
      if (len(words) > 0):
        firstlist.append(words)

  currentword = firstlist[random.randint(0,len(firstlist)-1)]
  #print(currentword)
  return currentword

#brings a list of selected corpus' sentences and selects last word from it
def lastword(selectedsentence):
  add = 0
  lastlist = list()
  for i in range(len(selectedsentence)-1):
    words = ""
    if selectedsentence[i] != '':
      #print("I ",i)
      for j in range(len(selectedsentence[i])-1):
        if (selectedsentence[i][-j-1] == " "):
          break
        else:
          words += selectedsentence[i][-j-1]
      if (len(words) > 0):
        words = reverse(words)
        lastlist.append(words)
  currentword = lastlist[random.randint(0,len(lastlist)-1)]
  #print(currentword)
  return currentword

def reverse(string):
    string = string[::-1]
    return string
```

# Unigram Model
The following function creates unigrams of a sentence/verse.

```python
def createunigram(selectwords):
  unigrams = []
  for i in range(len(selectwords)-1):
    w1 = selectwords[i]
    unigrams.append(tuple(w1))
    
  return unigrams
```

# Bigram Model
The following function creates bigrams of a sentence/verse.

```python

#CONTAINS ALL BIGRAMS OF SELECTED CORPUS
def createbigram(selectwords):
  bigrams = []
  for i in range(len(selectwords)-1):
    w1 = selectwords[i]
    w2 = selectwords[i+1]
    bigrams.append(tuple((w1, w2)))

  #len(bigrams)
  return bigrams

```

# Backward Bigram

The following function takes the corpus and processes it using the backwardbigram model and on the basis of probabilities, it generates new poetry. The UrdutoRoman() function then converts this poetry into Roman Urdu verse by verse.

> Ending word is slected from endword list and then its iterated backwardly



```python
def bibackstanza(biword,boolword = "DEFAULT"):
  currentword = ""
  lastrhymeword = ""
  bigrams = createbigram(poetrywords)

  for size in range(4):
    verse = []
    versesize = random.randint(7,10)
    if (size == 0):
      currentword = ""      #UNLIKE IN BIGRAM EACH SENTENCE HAS ITS OWN random words (bi continued from last)

      currentword = lastword(poetrysentence)
      lastrhymeword = currentword
    if (boolword != "DEFAULT" and size == 0):
      currentword = boolword
      lastrhymeword = currentword

    if (size != 0):       #USE UNIGRAM TO FIND RHYMING WORD BY MATCHING LAST LETTER OF THE WORDS
          unigrams = createunigram(poetrywords)
          similarwords = []
          for w in unigrams:
            if (w[-1] == lastrhymeword[-1]):
              similarwords.append(w)
          rhymecount = len(similarwords)-1
          rhymeword = random.randint(0,rhymecount)    #PICKS next word via probability
          rhymeindex = 0
          currentword = ""
          for i in similarwords:
            if rhymeindex == rhymeword:
              currentword = i
              #print(rhymeword,lastrhymeword[-1]," word ",i)
              break
            rhymeindex +=1
          lastrhymeword = ""
          for i in range(len(currentword)):
            lastrhymeword += currentword[i]

          currentword = lastrhymeword
      #print("Current word = ",currentword)

    #print("Current word = ",currentword)
    verse.append(currentword)
    #CONDITION IS 2 FOR BIGRAMS
    for v in range(versesize):
      currentbigram = []
      for w in bigrams:
        if (w[1] == currentword):
          currentbigram.append(w)

      wordcount = len(currentbigram)-1

      nextword = random.randint(0,wordcount)    #PICKS next word via probability
      index = 0
      for i in currentbigram:
        if index == nextword:
          currentword = i[0]
          break
        index +=1
      verse.append(currentword)
    verse.reverse()
    versestring = ""
    for i in verse:
      versestring += " " + i
    print(versestring)
    print(urdu2roman(versestring))
  print("\n")
  return None
```

# Bidirectional Bigram
The following function takes the corpus and processes it using the Bidirectional Bigram model and on the basis of probabilities, it generates new poetry. The UrdutoRoman() function then converts this poetry into Roman Urdu verse by verse.

```python
def bidrectional(biword,boolword = "DEFAULT"):
  currentword = ""
  backword = ""
  forwardword = ""
  #if biword == "f":
  bigrams = createbigram(poetrywords)
  for size in range(4):
    verse = ""
    backverse = []
    forwardverse = ""
    versesize = random.randint(7,10)
    #print(versesize)
    currentword = ""        #RANDOM WORD PICKED FOR ALL VERSES
    #if biword == "f":
    currentword = firstword(poetrysentence)

    if(boolword != "DEFAULT" and size == 0):
      currentword = boolword
    #print(currentword)
    forwardverse += " " + currentword
    for v in range(int(versesize/2)):
      forwardbigram = []
      backbigram = []
      for w in bigrams:
        if (size == 0):
          if (w[0] == currentword):
            forwardbigram.append(w)
          if (w[1] == currentword):
            backbigram.append(w)
        else:
          if (w[0] == forwardword):
            forwardbigram.append(w)
          if (w[1] == backword):
            backbigram.append(w)

      backwordcount = len(backbigram)-1
      forwardcount = len(forwardbigram)-1

      nextback = random.randint(0,backwordcount)    #PICKS next word via probability
      nextforward = random.randint(0,forwardcount)    #PICKS next word via probability
      indexback = 0
      indexforward = 0
     
      for i in backbigram:
        if indexback == nextback:
          backword = i[0]
          break
        indexback +=1
      for j in forwardbigram:
        if indexforward == nextforward:
          forwardword = j[1]
          break
        indexforward +=1
        #print(i[0], i[1])
      #print(currentword)  
      forwardverse += " " + forwardword
      backverse.append(backword)

    backverse.reverse()
    backversestring = ""
    for i in backverse:
      backversestring += " " + i
    verse += backversestring + forwardverse
    print(verse)
    print(urdu2roman(verse))
  print("\n")
  return None   
```

# Trigram Model
The following function creates trigrams of a sentence/verse.

```python
def createtrigram(selectwords):
  trigrams = list()
  for i in range(len(selectwords)-2):
    w1 = selectwords[i]
    w2 = selectwords[i+1]
    w3 = selectwords[i+2]
    trigrams.append(tuple((w1, w2, w3)))
  return trigrams
```
